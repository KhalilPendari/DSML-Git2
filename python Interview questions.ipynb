question 3 :
    Apply a custom function to each member, calculating bonus on their Age and Salary



The bonus should be 10% of the salary  for
individuals below the age of 30 and 15% for individuals  above or equal to the age 30

```python
def bonus (x):
  if x['Age']<30:
    x['Bonus']=x['Salary']*10/100
    return x

  if x['Age']>=30:
    x['Bonus']=x['Salary']*15/100
    return x
```

```python
df=df.apply(bonus,axis=1)
df
```
